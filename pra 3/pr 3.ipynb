```python
import pandas as pd
import numpy as np
```

```python
loan_df = pd.read_csv("loan_data_set.csv")
```

```python
print("Loan Dataset:")
display(loan_df.head())
```

```python
print("Dataset Information:")
loan_df.info()
```

```python
print("Missing Values:")
display(loan_df.isnull().sum())
```

```python
loan_df["LoanAmount"] = loan_df["LoanAmount"].fillna(loan_df["LoanAmount"].mean())
```

```python
loan_df["ApplicantIncome"] = loan_df["ApplicantIncome"].fillna(loan_df["ApplicantIncome"].mean())
```

```python
loan_df["Gender"] = loan_df["Gender"].fillna(loan_df["Gender"].mode()[0])
```

```python
loan_df["Education"] = loan_df["Education"].fillna(loan_df["Education"].mode()[0])
```

```python
print("Missing Values After Handling:")
display(loan_df.isnull().sum())
```

```python
print("Basic Statistical Summary:")
display(loan_df.describe())
```

```python
summary_gender = loan_df.groupby("Gender")["LoanAmount"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)
print("Summary Statistics of LoanAmount grouped by Gender:")
display(summary_gender)

```

```python
summary_education = loan_df.groupby("Education")["ApplicantIncome"].agg(
    Mean="mean",
    Median="median",
    Minimum="min",
    Maximum="max",
    Standard_Deviation="std"
)
print("Summary Statistics of ApplicantIncome grouped by Education:")
display(summary_education)

```

```python
loan_df["Gender_Numeric"] = loan_df["Gender"].map({"Male": 1, "Female": 0})
```

```python
gender_numeric_list = loan_df["Gender_Numeric"].tolist()
```

```python
print("Numeric list for Gender categorical variable:")
print(gender_numeric_list[:20])
```
