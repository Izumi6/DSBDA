```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("iris.csv")

print("First five rows of dataset:")
display(dataset.head())
```

```python
print("Shape of dataset:")
print(dataset.shape)
```

```python
print("Dataset Information:")
dataset.info()
```

```python
print("Missing Values:")
display(dataset.isnull().sum())
```

```python
print("Statistical Summary:")
display(dataset.describe())

```

```python
feature_info = pd.DataFrame({
    "Feature": ["sepal.length", "sepal.width", "petal.length", "petal.width", "variety"],
    "Description": [
        "Length of sepal",
        "Width of sepal",
        "Length of petal",
        "Width of petal",
        "Species/type of Iris flower"
    ],
    "Type": [
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Nominal / Categorical"
    ]
})

print("Feature Information:")
display(feature_info)
```

```python

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(dataset["sepal.length"], bins=15)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.subplot(2, 2, 2)
plt.hist(dataset["sepal.width"], bins=15)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")

plt.subplot(2, 2, 3)
plt.hist(dataset["petal.length"], bins=15)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

plt.subplot(2, 2, 4)
plt.hist(dataset["petal.width"], bins=15)
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(y=dataset["sepal.length"])
plt.title("Boxplot of Sepal Length")

plt.subplot(2, 2, 2)
sns.boxplot(y=dataset["sepal.width"])
plt.title("Boxplot of Sepal Width")

plt.subplot(2, 2, 3)
sns.boxplot(y=dataset["petal.length"])
plt.title("Boxplot of Petal Length")

plt.subplot(2, 2, 4)
sns.boxplot(y=dataset["petal.width"])
plt.title("Boxplot of Petal Width")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x="variety", y="sepal.length", data=dataset)
plt.title("Sepal Length by Variety")

plt.subplot(2, 2, 2)
sns.boxplot(x="variety", y="sepal.width", data=dataset)
plt.title("Sepal Width by Variety")

plt.subplot(2, 2, 3)
sns.boxplot(x="variety", y="petal.length", data=dataset)
plt.title("Petal Length by Variety")

plt.subplot(2, 2, 4)
sns.boxplot(x="variety", y="petal.width", data=dataset)
plt.title("Petal Width by Variety")

plt.tight_layout()
plt.show()
```
