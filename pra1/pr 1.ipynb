```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
df = pd.read_csv("titanic.csv")
```

```python
print("First five rows of dataset:")
display(df.head())
```

```python
print("Shape of dataset:")
print(df.shape)
```

```python
print("Dataset information:")
df.info()
```

```python
print("Missing values before handling:")
display(df.isnull().sum())
```

```python
print("Statistical summary:")
display(df.describe())

```

```python
print("Data types before formatting:")
display(df.dtypes)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
print("Dataset after filling missing Age values:")
display(df.head())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print("Dataset after filling missing Embarked values:")
display(df.head())
```

```python
df.drop("Cabin", axis=1, inplace=True)
print("Dataset after dropping Cabin column:")
display(df.head())
```

```python
print("Missing values after handling:")
display(df.isnull().sum())
```

```python
df["PassengerId"] = df["PassengerId"].astype(int)
df["Survived"] = df["Survived"].astype(int)
df["Pclass"] = df["Pclass"].astype(int)
df["Age"] = df["Age"].astype(float)
df["Fare"] = df["Fare"].astype(float)
print("Dataset after type conversion:")
display(df.head())
print("Data types after formatting:")
display(df.dtypes)


```

```python
scaler = MinMaxScaler()
df[["Age", "Fare"]] = scaler.fit_transform(df[["Age", "Fare"]])
print("Dataset after normalization of Age and Fare:")
display(df.head())
```

```python
le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
print("Dataset after Label Encoding of Sex column:")
display(df.head())
```

```python
df = pd.get_dummies(df, columns=["Embarked"])
print("Dataset after One-Hot Encoding of Embarked column:")
display(df.head())
```

```python
print("Final missing values:")
display(df.isnull().sum())
print("Final data types:")
display(df.dtypes)

```

```python
print("Final shape of dataset:")
print(df.shape)
```

```python
print("Final dataset:")
display(df.head())
```
