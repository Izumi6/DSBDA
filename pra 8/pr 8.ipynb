```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("titanic.csv")

print("First five rows of Titanic dataset:")
display(titanic.head())
```

```python
print("Shape of dataset:")
print(titanic.shape)
```

```python
print("Dataset Information:")
titanic.info()
```

```python
print("Missing Values:")
display(titanic.isnull().sum())

```

```python
print("Statistical Summary:")
display(titanic.describe())
```

```python
sns.countplot(x="Survived", data=titanic)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=titanic)
plt.title("Gender-wise Survival Count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.title("Passenger Class-wise Survival Count")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(titanic["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

```

```python
sns.histplot(titanic["Fare"], bins=30, kde=True)
plt.title("Fare Distribution of Passengers")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Fare", data=titanic)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", data=titanic)
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
